## 시각화 통합 분류표

"변수가 몇 개인가?" × "무엇을 보여주고 싶은가?"를 교차하여 차트를 선택합니다.

| | 비교 (Comparison) | 분포 (Distribution) | 구성 (Composition) | 관계 (Relationship) |
|---|---|---|---|---|
| **단변량** | 막대그래프 | 히스토그램, KDE, 박스플롯, 바이올린 | 파이, 도넛, 트리맵, 와플 | - |
| **이변량** | 그룹별 박스플롯 ¹, 그룹별 바이올린 ¹, 그룹별 히스토그램 ¹, 선 그래프, 영역 그래프 ², 다중축 | 그룹별 박스플롯 ¹, 그룹별 바이올린 ¹, 그룹별 히스토그램 ¹ | 누적 막대, 모자이크, 영역 그래프 ² | 산점도, 히트맵 |
| **다변량** | 패싯, 레이더 | 페어플롯 ³ | 생키 | 버블, 페어플롯 ³ |
| **특수 목적** | 간트 | - | 게이지 | - |

> **¹** 그룹별 박스플롯·바이올린·히스토그램: 그룹 간 **비교**이면서 동시에 각 그룹의 **분포**를 보여줌 \
> **²** 영역 그래프: 추세 **비교**이면서 누적 시 **구성** 비율도 표현 \
> **³** 페어플롯: 대각선은 변수별 **분포**, 비대각선은 변수 쌍의 **관계**를 동시에 표현

## 시각화 상세

### 단변량 (변수 1개)

| 데이터 유형 | 시각화 | 용도 | 언제 쓰나? |
|------------|--------|------|-----------|
| **수치형 (연속)** | 히스토그램 | 분포 형태 확인 (어디에 몰려 있는지) | 데이터 분포의 전체 모양 파악 |
| | KDE (커널 밀도) | 부드러운 분포 곡선 | 히스토그램보다 매끄러운 분포 비교 |
| | 박스플롯 | 중앙값, 사분위, 이상치 요약 | 이상치 탐지, 분포 범위 파악 |
| | 바이올린 플롯 | 박스플롯 + 분포 형태 | 분포의 봉우리, 치우침까지 확인 |
| **범주형** | 막대그래프 (Bar) | 범주별 빈도/크기 비교 | 항목 간 크기 비교 |
| | 파이차트 (Pie) | 전체 대비 비율 | 항목이 적고(≤5) 비율 강조할 때 |
| | 도넛 차트 (Donut) | 파이차트 변형, 중앙에 정보 표시 | 핵심 수치를 중앙에 강조할 때 |
| | 트리맵 (Treemap) | 계층적 비율을 면적으로 표현 | 항목이 많거나 계층 구조가 있을 때 |
| | 와플 차트 (Waffle) | 비율을 격자로 직관 표현 | 파이차트 대안, 퍼센트 강조할 때 |

### 이변량 (변수 2개)

| 데이터 조합 | 시각화 | 용도 | 언제 쓰나? |
|------------|--------|------|-----------|
| **수치 × 수치** | 산점도 (Scatter) | 두 변수의 관계 확인 | 상관관계, 군집, 이상치 패턴 확인 |
| | 히트맵 (상관행렬) | 변수 쌍별 상관관계를 행렬로 요약 | 변수가 많을 때 전체 관계를 색깔로 파악 |
| **수치 × 범주** | 그룹별 박스플롯 | 그룹 간 분포 비교 | 남녀별 키, 등급별 요금 등 |
| | 그룹별 바이올린 | 그룹 간 분포 + 형태 비교 | 분포 형태까지 세밀하게 비교할 때 |
| | 그룹별 히스토그램 | 그룹 분포 겹쳐 비교 | 분포의 겹침/분리 정도 확인 |
| **시간 × 수치** | 선 그래프 (Line) | 시간에 따른 변화 추세 | 매출 추이, 기온 변화, 주가 등 |
| | 영역 그래프 (Area) | 추세 + 누적 크기 강조 | 카테고리별 매출 누적 추이 |
| | 다중 축 차트 (Multi-Axis) | 스케일이 다른 변수를 겹쳐 비교 | 매출(억) vs 고객수(명) 동시 비교 |
| **범주 × 범주** | 누적 막대그래프 | 교차 빈도/비율 | 성별×생존, 등급×탑승항 등 |
| | 모자이크 플롯 | 교차 비율 (면적) | 두 범주 변수의 관계를 면적으로 표현 |

### 다변량 (변수 3개 이상)

| 시각화 | 용도 | 언제 쓰나? |
|--------|------|-----------|
| **페어플롯 (Pair Plot)** | 모든 변수 쌍 관계를 한눈에 | EDA 초기 전체 변수 탐색 |
| **버블 차트 (Bubble)** | 산점도 + 3번째 변수를 크기로 | 국가별 GDP×기대수명×인구 등 |
| **패싯 그리드 (Facet)** | 기존 차트를 범주 조합별로 분할 배치 | 성별×등급 등 그룹 조합별 패턴을 나란히 비교 |
| **레이더 차트 (Radar)** | 그룹별 다항목 프로필 비교 | 군집분석 결과, 고객 유형별 특성 비교 |
| **생키 다이어그램 (Sankey)** | 단계별 흐름·전환 경로 | 유입경로→전환→이탈 등 흐름 분석 |

### 특수 목적

| 시각화 | 용도 | 언제 쓰나? |
|--------|------|-----------|
| **게이지 차트 (Gauge)** | 목표 대비 현재 달성률 | KPI 대시보드, 실적 현황판 |
| **간트 차트 (Gantt)** | 작업별 기간과 일정 | 프로젝트 관리, 일정 계획 |

```python
# 필수 라이브러리 Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import squarify
from matplotlib.patches import FancyBboxPatch
from statsmodels.graphics.mosaicplot import mosaic
import warnings
import platform

warnings.filterwarnings('ignore')

# seaborn 테마 설정
sns.set_theme(style="whitegrid", palette="muted", font_scale=1.1)

# 운영체제별 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rcParams['font.family'] = 'Malgun Gothic'
elif platform.system() == 'Darwin':
    plt.rcParams['font.family'] = 'AppleGothic'
else:
    plt.rcParams['font.family'] = 'NanumGothic'

plt.rcParams['axes.unicode_minus'] = False


# 재현성
np.random.seed(42)

# 샘플 데이터 준비
iris = load_iris()
df = pd.DataFrame(iris.data, columns=['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비'])
df['품종'] = pd.Categorical([iris.target_names[t] for t in iris.target])

# 타이타닉 유사 데이터 (범주×범주 시각화용)
titanic = sns.load_dataset('titanic')

# 시계열 데이터 (선 그래프용)
dates = pd.date_range('2024-01', periods=12, freq='ME')
sales = pd.DataFrame({
    '월': dates,
    '매출': [120, 135, 148, 162, 155, 178, 195, 210, 198, 225, 240, 260],
    '비용': [80, 85, 90, 95, 92, 100, 110, 115, 108, 120, 125, 135]
})

print("=" * 60)
print("데이터 시각화 가이드 - 실습")
print("라이브러리 및 데이터 로드 완료!")
print("=" * 60)
```

  ## Part 1: 단변량 분석 (변수 1개: 분포 파악)



  단변량 시각화는 **하나의 변수의 분포**를 파악하는 데 사용합니다.

  "데이터가 어디에 몰려 있는지", "이상치가 있는지", "어떤 항목이 많은지"를 확인합니다.

```python
# =========================================================================
# 1-1. 수치형: 히스토그램 + KDE
# =========================================================================
# 히스토그램: 구간(bin)별 빈도를 막대로 표현
# KDE: 히스토그램을 부드럽게 연결한 확률밀도 곡선
print("\n[1-1] 히스토그램 + KDE")

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# (a) 히스토그램만
sns.histplot(df['꽃잎길이'], bins=20, color='steelblue', ax=axes[0])
axes[0].set_title('(a) 히스토그램', fontweight='bold')
axes[0].set_xlabel('꽃잎 길이 (cm)')
axes[0].set_ylabel('빈도')

# (b) KDE만
sns.kdeplot(df['꽃잎길이'], fill=True, color='coral', ax=axes[1])
axes[1].set_title('(b) KDE (커널 밀도)', fontweight='bold')
axes[1].set_xlabel('꽃잎 길이 (cm)')
axes[1].set_ylabel('밀도')

# (c) 히스토그램 + KDE 겹치기
sns.histplot(df['꽃잎길이'], bins=20, kde=True, color='steelblue', ax=axes[2])
axes[2].set_title('(c) 히스토그램 + KDE', fontweight='bold')
axes[2].set_xlabel('꽃잎 길이 (cm)')
axes[2].set_ylabel('빈도')

plt.suptitle('수치형 단변량: 히스토그램 & KDE', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 1-2. 수치형: 박스플롯 + 바이올린 플롯
# =========================================================================
# 박스플롯: 중앙값, Q1, Q3, 이상치를 한눈에
# 바이올린: 박스플롯 + 분포 형태(밀도)까지 표현
print("\n[1-2] 박스플롯 + 바이올린 플롯")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# (a) 박스플롯
sns.boxplot(y=df['꽃잎길이'], color='lightblue', width=0.4, ax=axes[0])
axes[0].set_title('(a) 박스플롯', fontweight='bold')
axes[0].set_ylabel('꽃잎 길이 (cm)')

# (b) 바이올린 플롯
sns.violinplot(y=df['꽃잎길이'], color='lightsalmon', inner='quartile', ax=axes[1])
axes[1].set_title('(b) 바이올린 플롯', fontweight='bold')
axes[1].set_ylabel('꽃잎 길이 (cm)')

plt.suptitle('수치형 단변량: 박스플롯 & 바이올린', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 1-3. 범주형: 막대그래프 + 파이차트
# =========================================================================
# 타이타닉 등급 데이터 사용 (불균형 → 막대/파이 비교 효과 명확)
print("\n[1-3] 막대그래프 + 파이차트")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# (a) 막대그래프
class_counts = titanic['class'].value_counts()
sns.barplot(x=class_counts.index, y=class_counts.values,
            palette='Set2', ax=axes[0])
axes[0].set_title('(a) 막대그래프', fontweight='bold')
axes[0].set_xlabel('객실 등급')
axes[0].set_ylabel('인원수')
for i, v in enumerate(class_counts.values):
    axes[0].text(i, v + 3, str(v), ha='center', fontweight='bold')

# (b) 파이차트
colors = sns.color_palette('Set2', len(class_counts))
axes[1].pie(class_counts.values, labels=class_counts.index,
            autopct='%1.1f%%', colors=colors, startangle=90,
            wedgeprops={'edgecolor': 'white', 'linewidth': 2})
axes[1].set_title('(b) 파이차트', fontweight='bold')

plt.suptitle('범주형 단변량: 막대그래프 & 파이차트', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 1-4. 범주형: 트리맵
# =========================================================================
# 면적으로 비율을 표현 → 항목이 많거나 계층 구조가 있을 때 적합
# pip install squarify / uv add squarify
print("\n[1-4] 트리맵 (Treemap)")

fig, ax = plt.subplots(figsize=(10, 6))

# 타이타닉 탑승항별 인원수 (불균형이 뚜렷)
embark_counts = titanic['embark_town'].value_counts().dropna()
colors = sns.color_palette('Set2', len(embark_counts))
squarify.plot(sizes=embark_counts.values,
              label=[f'{name}\n({count}명, {count/embark_counts.sum()*100:.1f}%)'
                     for name, count in embark_counts.items()],
              color=colors, alpha=0.85, edgecolor='white', linewidth=3,
              text_kwargs={'fontsize': 13, 'fontweight': 'bold'}, ax=ax)
ax.set_title('트리맵: 탑승항별 승객 비율', fontsize=14, fontweight='bold')
ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 1-5. 범주형: 와플 차트 (matplotlib로 직접 구현)
# =========================================================================
# 비율을 격자로 직관 표현 → 파이차트 대안, 퍼센트 강조
print("\n[1-5] 와플 차트 (Waffle Chart)")

fig, ax = plt.subplots(figsize=(10, 5))

# 타이타닉 등급별 비율을 10×10 격자로 표현
class_pct = (class_counts / class_counts.sum() * 100).round().astype(int)
# 반올림 합이 100이 되도록 보정
diff = 100 - class_pct.sum()
class_pct.iloc[0] += diff

colors_map = dict(zip(class_counts.index, sns.color_palette('Set2', len(class_counts))))
grid = []
for cls, pct in class_pct.items():
    grid.extend([cls] * pct)

for i in range(10):
    for j in range(10):
        idx = i * 10 + j
        cls = grid[idx]
        rect = FancyBboxPatch((j, 9 - i), 0.9, 0.9, boxstyle='round,pad=0.05',
                               facecolor=colors_map[cls], edgecolor='white', linewidth=2)
        ax.add_patch(rect)

ax.set_xlim(-0.2, 10.2)
ax.set_ylim(-0.2, 10.2)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('와플 차트: 객실 등급별 비율 (1칸 = 1%)', fontsize=14, fontweight='bold')

# 범례
legend_patches = [plt.Rectangle((0, 0), 1, 1, facecolor=colors_map[cls])
                  for cls in class_counts.index]
ax.legend(legend_patches,
          [f'{cls} ({class_pct[cls]}%)' for cls in class_counts.index],
          loc='center left', bbox_to_anchor=(1.02, 0.5), fontsize=11)

plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 1-6. 범주형: 도넛 차트
# =========================================================================
# 파이차트 변형으로 중앙에 핵심 수치를 강조할 수 있음
print("\n[1-6] 도넛 차트 (Donut Chart)")

fig, ax = plt.subplots(figsize=(7, 7))

class_counts = titanic['class'].value_counts()
colors = sns.color_palette('Set2', len(class_counts))

# 도넛: wedgeprops의 width로 구멍 크기 조절
wedges, texts, autotexts = ax.pie(
    class_counts.values, labels=class_counts.index,
    autopct='%1.1f%%', colors=colors, startangle=90,
    wedgeprops={'width': 0.4, 'edgecolor': 'white', 'linewidth': 2},
    pctdistance=0.8)

for t in autotexts:
    t.set_fontweight('bold')

# 중앙에 핵심 수치 표시
ax.text(0, 0, f'총 {class_counts.sum()}명', ha='center', va='center',
        fontsize=18, fontweight='bold', color='#333333')
ax.set_title('도넛 차트: 객실 등급별 비율', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

  ## Part 2: 이변량 분석 (변수 2개: 관계 파악)



  이변량 시각화는 **두 변수 사이의 관계**를 파악하는 데 사용합니다.

  데이터 조합(수치×수치, 수치×범주, 시간×수치, 범주×범주)에 따라 적합한 차트가 다릅니다.

```python
# =========================================================================
# 2-1. 수치 × 수치: 산점도
# =========================================================================
print("\n[2-1] 산점도 (Scatter Plot)")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# (a) 기본 산점도
sns.scatterplot(data=df, x='꽃잎길이', y='꽃잎너비', ax=axes[0],
                color='steelblue', alpha=0.7, s=60)
axes[0].set_title('(a) 기본 산점도', fontweight='bold')

# (b) 그룹별 산점도 (hue 활용)
sns.scatterplot(data=df, x='꽃잎길이', y='꽃잎너비', hue='품종',
                palette='Set2', alpha=0.7, s=60, ax=axes[1])
axes[1].set_title('(b) 품종별 산점도', fontweight='bold')
axes[1].legend(title='품종')

plt.suptitle('수치 × 수치: 산점도', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 3-4. 레이더 차트 (Radar Chart)
# =========================================================================
# 여러 항목(변수)의 값을 방사형으로 배치하여 그룹 간 프로필을 비교
# 군집분석 결과, 고객 유형별 특성 비교 등에 활용
print("\n[3-4] 레이더 차트 (Radar Chart)")

# 품종별 평균으로 프로필 생성
categories = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비']
species_means = df.groupby('품종')[categories].mean()

# 0~1 정규화 (비교를 위해)
species_norm = (species_means - species_means.min()) / (species_means.max() - species_means.min())

# 레이더 차트 그리기
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(7, 7), subplot_kw={'polar': True})
colors = sns.color_palette('Set2', len(species_norm))

for i, (species, row) in enumerate(species_norm.iterrows()):
    values = row.tolist() + [row.tolist()[0]]
    ax.plot(angles, values, linewidth=2, label=species, color=colors[i])
    ax.fill(angles, values, alpha=0.15, color=colors[i])

ax.set_thetagrids(np.degrees(angles[:-1]), categories)
ax.set_ylim(0, 1.1)
ax.set_title('레이더 차트: 품종별 특성 프로필', fontsize=14,
             fontweight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1), title='품종')

plt.tight_layout()
plt.show()
```

  ## Part 4: 특수 목적 차트



  대시보드, 프로젝트 관리 등 특정 상황에서 사용하는 차트입니다.

```python
# =========================================================================
# 4-1. 게이지 차트 (Gauge Chart)
# =========================================================================
# 목표 대비 달성률을 반원형으로 표현
# KPI 대시보드에서 "현재 상태"를 한눈에 보여줄 때 사용
print("\n[4-1] 게이지 차트 (Gauge Chart)")

fig, axes = plt.subplots(1, 3, figsize=(16, 5))

kpis = [
    {'name': '매출 달성률', 'value': 78, 'color': '#2ecc71'},
    {'name': '고객 만족도', 'value': 92, 'color': '#3498db'},
    {'name': '불량률', 'value': 15, 'color': '#e74c3c'},
]

for ax, kpi in zip(axes, kpis):
    # 배경 반원 (회색)
    theta_bg = np.linspace(np.pi, 0, 100)
    ax.plot(np.cos(theta_bg), np.sin(theta_bg), color='#e0e0e0', linewidth=20,
            solid_capstyle='round')

    # 달성률 반원 (색상)
    ratio = kpi['value'] / 100
    theta_val = np.linspace(np.pi, np.pi - (np.pi * ratio), 100)
    ax.plot(np.cos(theta_val), np.sin(theta_val), color=kpi['color'], linewidth=20,
            solid_capstyle='round')

    # 중앙 수치
    ax.text(0, -0.1, f"{kpi['value']}%", ha='center', va='center',
            fontsize=28, fontweight='bold', color=kpi['color'])
    ax.set_title(kpi['name'], fontsize=13, fontweight='bold', pad=15)

    ax.set_xlim(-1.3, 1.3)
    ax.set_ylim(-0.5, 1.3)
    ax.set_aspect('equal')
    ax.axis('off')

plt.suptitle('게이지 차트: KPI 대시보드', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# =========================================================================
# 4-2. 간트 차트 (Gantt Chart)
# =========================================================================
# 프로젝트 작업별 시작/종료 기간을 가로 막대로 표현
print("\n[4-2] 간트 차트 (Gantt Chart)")

fig, ax = plt.subplots(figsize=(12, 5))

# 가상 프로젝트 일정
tasks = pd.DataFrame({
    '작업': ['데이터 수집', '전처리', 'EDA', '모델링', '평가', '보고서 작성'],
    '시작': [0, 3, 5, 8, 12, 14],
    '기간': [4, 3, 4, 5, 3, 3]
})

colors = sns.color_palette('Set2', len(tasks))

for i, (_, row) in enumerate(tasks.iterrows()):
    ax.barh(i, row['기간'], left=row['시작'], height=0.6,
            color=colors[i], edgecolor='white', linewidth=1.5)
    ax.text(row['시작'] + row['기간'] / 2, i, f"{row['기간']}일",
            ha='center', va='center', fontweight='bold', fontsize=10, color='white')

ax.set_yticks(range(len(tasks)))
ax.set_yticklabels(tasks['작업'])
ax.set_xlabel('일차', fontsize=11)
ax.set_title('간트 차트: 데이터 분석 프로젝트 일정', fontsize=14, fontweight='bold')
ax.invert_yaxis()
ax.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```

  ## 시각화 실전 원칙



  차트를 만든 뒤 아래 항목을 점검하면 흔한 실수를 줄일 수 있습니다.



  ### 축과 데이터

  - **시간 축**: 시간은 항상 가로축에 배치하고, 왼쪽→오른쪽으로 진행. 중간 값을 건너뛰지 않는다

  - **비례 표현**: 막대/영역/버블의 크기는 실제 수치에 정비례해야 한다

  - **0 기준선**: 막대 차트는 반드시 0에서 시작 (잘린 축은 왜곡을 유발)

  - **정렬**: 막대 차트는 알파벳순이 아닌 **값 기준** 오름/내림차순으로 정렬



  ### 색상

  - 한 차트에 **6개 이상의 색상을 사용하지 않는다**

  - 같은 변수의 시간 변화: **같은 색의 다른 명도**(밝은→어두운)로 표현

  - 다른 범주: **서로 다른 색상** 사용

  - 시리즈 전체에서 **동일한 색상 팔레트** 유지 → 일관성 확보

  - 흑백 인쇄 시에도 구분되는지, 색맹 접근성도 확인



  ### 라벨과 범례

  - 범주가 1개뿐이면 **범례 불필요**

  - 가능하면 선/막대에 **직접 라벨** 부착 (범례 간접 조회 최소화)

  - 인플레이션 등 장기 금액 데이터는 **실질 가치로 보정**



  ### 복잡성 관리

  - 한 차트에 너무 많은 정보를 넣지 않는다

  - 필요하면 **차트를 분할**하거나 강조 표시로 단순화

  - 핵심 메시지 하나에 집중하는 차트가 가장 효과적
